# Results from neurocomputing

### Data description

For our training purposes, we contaminated data by R function **contaminate** from **simFrame** package. This function try to make distribution $F = (1-\epsilon)G + \epsilon H$ where $G$, is the distbution of noncaminated data, and $H$ is distribution of contamination.

For more information about process which generates outliers, check 
[this refrence manual](https://personal.eur.nl/alfons/publications/CDAM2010.pdf).

I create new datasets with outlying observations, in R. But whole training was done in Python with help of Keras and Tensorflow.

Data Sets we used during our training was Auto Mpg dataset, Boston Housing Dataset, California Housing Dataset and artificaly generated data which can be seen in next section.



### Artificaly data results

![dsds](https://raw.githubusercontent.com/tomasj12/test_test/master/Rplot.png)







On this data we did 10-fold Cross Validation. After CV, we check how our models are good by metrics:

$$MSE = \frac{1}{n}\sum_{i = 1}^{n}u_{i}^{2} \quad TMSE(\alpha) = \frac{1}{h}\sum_{i = 1}^{h}u_{(i)}^{2}  .$$

We create and contaminated data with various values of $\epsilon \in (0,4]$. All epsilon values are:


```r
0.0 0.05 0.06 0.07 0.08 0.09 0.10 0.11 0.12 0.13 0.14 0.15 0.16 0.17 0.18 0.19 0.20 0.21 0.22 0.23 0.24 0.25
0.26 0.27 0.28 0.29 0.30 0.31 0.32 0.33 0.34 0.35
```

The main formula of data generate is based on formula: $$ Y = \sin(10x) + c $$, where $c \sim N(0,\frac{1}{2})$ and $x \in [-1,1]$. Also we for every dataset, and for current contamination, set for those data, which should be contaminated, extreme constant. (for exmaple 5,6, or 20). 

On choosed set of artifically generated datasets, we applied our model.
